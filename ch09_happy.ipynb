```python
happy_safe = pd.read_excel('ch09_data/대한민국행복지도_안전.xlsx')
happy_safe.head()
```

```python
happy_relation = pd.read_excel('ch09_data/대한민국행복지도_관계및사회참여.xlsx')
happy_relation.head()
```

```python
happy_leisure = pd.read_excel('ch09_data/대한민국행복지도_여가.xlsx')
happy_leisure.head()
```

```python
happy_environ = pd.read_excel('ch09_data/대한민국행복지도_환경.xlsx')
happy_environ.head()
```

```python
happy_health = pd.read_excel('ch09_data/대한민국행복지도_건강.xlsx')
happy_health.head()
```

# 2) 데이터 병합

```python
city = happy_life['시도'].unique()
city
```

```python
# 리스트로 변환
city = list(city)
city
```

```python
# 데이터프레임으로 생성
happy_merge = pd.DataFrame({
    '시도': city
})
happy_merge.head()
```

## 시도별 삶의 만족도의 평균

```python
life = happy_life['삶의 만족도'].groupby(by=happy_life['시도']).mean()
life
```

```python
# '시도'를 중심으로 병합한다.
happy_merge = pd.merge(happy_merge, life, on='시도')
happy_merge.head()
```

## 시도별 건강의 평균

```python
health = happy_health['평균'].groupby(happy_health['시도']).mean()
health
```

```python
happy_merge = pd.merge(happy_merge, health, on='시도')
happy_merge.head()
```

```python
happy_merge = happy_merge.drop(columns=['삶의 만족도_y'])

happy_merge = happy_merge.rename(columns={'삶의 만족도_x': '삶의 만족도'})

happy_merge.head()
```

```python
happy_merge = happy_merge.rename(columns={'평균': '건강'})

happy_merge.head()
```

## 시도별 안전의 평균

```python
safe = happy_safe['평균'].groupby(happy_safe['시도']).mean()
safe
```

```python
happy_merge = pd.merge(happy_merge, safe.rename('안전'), on='시도')
happy_merge.head()
```

## 시도별 환경의 평균

```python
environ = happy_environ['평균'].groupby(happy_environ['시도']).mean()
happy_merge = pd.merge(happy_merge, environ.rename('환경'), on='시도')
happy_merge.head()
```

## 시도별 경제의 평균

```python
econo = happy_econo['평균'].groupby(happy_econo['시도']).mean()
happy_merge = pd.merge(happy_merge, econo.rename('경제'), on='시도')
happy_merge.head()
```

## 시도별 관계및사회참여의 평균

```python
relation = happy_relation['평균'].groupby(happy_relation['시도']).mean()
happy_merge = pd.merge(happy_merge, relation.rename('관계및사회참여'), on='시도')
happy_merge.head()
```

## 시도별 여가의 평균

```python
lesiure = happy_leisure['평균'].groupby(happy_leisure['시도']).mean()
happy_merge = pd.merge(happy_merge, lesiure.rename('여가'), on='시도')
happy_merge.head()
```

## 시도별 교육의 형균

```python
edu = happy_edu['평균'].groupby(happy_edu['시도']).mean()
happy_merge = pd.merge(happy_merge, edu.rename('교육'), on='시도')
happy_merge.head()
```

# 4) 시각화

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import seaborn as sns

# 한글 폰트 설정
font_path = 'C:\Windows\Fonts\malgun.ttf'
font_name = fm.FontProperties(fname=font_path, size=12).get_name()

plt.rc('font', family = font_name)
```

```python
plt.figure(figsize=(15, 6))     # 가로 15인치, 세로 6인치
items = list(happy_merge.columns[1:9])
for a in items:
    chartdata = happy_merge[a]
    plt.plot(chartdata, marker='^', label=a)

plt.xlabel('시도') # x축 제목
plt.ylabel('수치')
plt.xticks(range(17), happy_merge['시도'], fontsize=8)
plt.title('대한민국 행복지수') # 차트 제목
plt.legend() # 범례
plt.grid()
plt.show()
```
