### 라이브러리 불러오기

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder # 사이킷런 - 전처리 모듈 -> 레이블인코더, 원핫인코더
```

### 데이터 불러오기

```python
CSV_PATH = 'Clean_Dataset.csv'
df_origin = pd.read_csv(CSV_PATH)

df_origin.head()
```

```python
pd.set_option('display.max_columns', None) # 컬럼이 많아도 잘리지 않고 다 보이게 설정
```

```python
df_origin = df_origin.drop('Unnamed: 0', axis=1)  # 컬럼(Unnamed: 0) 삭제 
df_origin.head()
```

### 레이블 인코딩
- 범주형 데이터(문자열 데이터)에 숫자를 순서대로 매기는 방식
- 판다스와 사이킷런 모두 활용 가능

### 판다스에서 레이블 인코딩하기 - pd.factorize()

```python
df = df_origin.copy() # 데이터프레임 복사
df.info()
```

```python
df['airline'].value_counts()
```

```python
pd.factorize(df['airline'])
```

```python
pd.factorize(df['airline'])[0]  # 1차원
```

```python
df.shape  # 행 개수, 열 개수
```

### .reshape(-1, 1) : 차원을 바꿔주는 함수
- 1차원 배열을 (행 수, 1) 형태의 2차원으로 바꿔서 데이터프레임의 새 컬럼으로 대입하기 쉽게 한다.

```python
pd.factorize(df['airline'])[0].reshape(-1, 1)  # 2차원
```

```python
df['label_encoding'] = pd.factorize(df['airline'])[0].reshape(-1, 1)

df.head()
```

### 사이킷런으로 레이블 인코딩하기 - LabelEncoder (알파벳 순서로)

### fit_transform()
- fit : 범주(문자열) 학습
- transform : 숫자 변환
- 학습하고 숫자변환을 한 번에 수행

```python
le = LabelEncoder()
le.fit_transform(df['airline'])
```

```python
# pd.set_option('display.max_columns', None) # 컬럼이 많아도 잘리지 않고 다 보이게 설정
```

```python
df['airline_Label_Encoder'] = le.fit_transform(df['airline'])

df.head()
```

#### 레이블 인코딩 역변환(디코딩)하기 - .inverse_transform()

```python
le.inverse_transform(df['airline_Label_Encoder'])
```

```python
df['decoded'] = le.inverse_transform(df['airline_Label_Encoder'])

df.head()
```

### 원핫 인코딩
- 레이블 인코딩 시 문제점이 의도치 않게 숫자 크기 정보가 생겨서 이를 크기 순서로 오해할 경우가 있다.
- 이 문제를 해결하기 위해 원핫 인코딩이 등장
- 각 범주(문자열)를 독립된 0 또는 1 컬럼으로 분리한다.
- 범주 개수가 많아지면 컬럼 수가 급증하는 트레이드오프가 있다.
- 범주 개수가 적으면 원핫 인코딩을 추천, 많을 때는 신중히 판단을 한다.

### 판다스에서 원핫 인코딩하기 - pd.get_dummies()

```python
df = df_origin.copy()
df.head()
```

```python
df['class'].value_counts()
```

```python
# class 컬럼을 원핫 인코딩
pd.get_dummies(df['class'])
```

```python
df = pd.get_dummies(df['class'])  # 'class'속성만 원핫 인코딩
df.head()
```

```python
df2 = df_origin.copy()
```

```python
df2 = pd.get_dummies(df2)  # df2라는 데이터프레임 전체(범주형 컬럼 다 해당) 원핫 인코딩
df2.head()
```

### 사이킷런으로 원핫 인코딩하기 - OneHotEncoder 클래스

```python
df = df_origin.copy()

oh = OneHotEncoder()

# fit_transform()에는 2차원 배열이 들어가야 한다. 
df['class'].values
```

```python
df['class'].values.reshape(-1, 1)
```

```python
oh.fit_transform(df['class'].values.reshape(-1, 1)).toarray()
```

- OntHotEncoder는 배열을 반환(array) 
- 왜 바로 데이터프레임에 반영되지 않는가??
    - 머신러닝 파이프라인에 바로 연결해서 train/test 데이터에 동일한 인코딩 규칙을 일관되게 적용하기 위해 
        설계된 도구이기 때문
    - 단순 EDA/전처리용으로는 get_dummies가 훨씬 간편하다.
    - 실제 머신러닝 모델링에서는 OneHotEncoder를 사용하는 경우도 많다.
